```python
import os
from pinecone import ServerlessSpec

cloud = os.environ.get('PINECONE_CLOUD') or 'aws'
region = os.environ.get('PINECONE_REGION') or 'us-east-1'

spec = ServerlessSpec(cloud=cloud, region=region)
```

```python

import pdfplumber

# Load the PDF file
pdf_file_path = "/content/system.pdf"
with pdfplumber.open(pdf_file_path) as pdf:
    document_text = ""
    for page in pdf.pages:
        document_text += page.extract_text()


```

```python
print(document_text)
```

```python
pinecone_client = pinecone.Pinecone(api_key="7ba597c6-712f-a26a-9f5993ed1882", environment="us-west1-gcp-free") #Use your Pinecone and Cohere api keys
co = cohere.Client('1esBQiIg4ILGYf5FEslCwo5S9P2b44Ksdxv')

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(document_text)

index_name = 'document-qa-index'
if index_name not in pinecone_client.list_indexes().names():
    pinecone_client.create_index(index_name, dimension=embeddings.shape[1],spec=spec)

# Connect to the index
index = pinecone_client.Index(index_name)

# Insert document embeddings into Pinecone
# Convert embedding to a list
index.upsert([(f'doc_0', embeddings.tolist(), {'text': document_text})])

print("Document embeddings inserted into Pinecone!")
```

```python
def retrieve_and_answer(question):
    # Encode the question
    query_embed = model.encode([question])

    # Query Pinecone for the most relevant documents
    # Convert query_embed to a list
    result = index.query(vector=query_embed.tolist(), top_k=3, include_metadata=True)

    # Extract relevant document
    retrieved_docs = [doc['id'] for doc in result['matches']]

    # Use Cohere to generate a response from the relevant document
    prompt = f"Answer the question '{question}' based on the following context: {retrieved_docs}"
    response = co.generate(
        model='command-r-plus',
        prompt=prompt,
        max_tokens=200,
        temperature=0.5,
        k=0
    )
    return response.generations[0].text

# Example query
question = "limitaions of decsion tree regressgor,use line break"
answer = retrieve_and_answer(question)
print(answer)
```
